```python
import pandas as pd

df=pd.read_csv("athlete_events.csv")

print(df.head())

#PARTE 1

#Veterano en ganar medalla

#Filtro datos de persona que ganaron medalla
df_medallas=df[df.Medal.notnull()]
index=df_medallas["Age"].idxmax()
print("Veterano en recibir medalla: ", df.iloc[index])


#Atleta mas joven en recibir medalla oro

#Filtro datos de personas que ganaron medalla de oro
df_oro=df[df["Medal"]=="Gold"]

index2=df_oro["Age"].idxmin()
print("Persona mas joven en recibir medalla de oro: ",df.iloc[index2])


#Atleta mas condecorado
index_atleta=df_medallas["Name"].value_counts().idxmax()
print("Atleta mas condecorado: ", index_atleta)

condecorado=df[df["Name"]==index_atleta]
condecorado.to_csv("condecorado.csv")


#PARTE 2

#Competidor mas veterano de los NOC´s MEX, COL y ARG
NOC2=["MEX","COL","ARG"]
df_veterano=df_medallas[df_medallas["NOC"]. isin(NOC2) ]

index3=df_veterano["Age"].idxmax()
print("Competidor mas veterano en ganar medalla de los NOC´s MEX, COL y ARG: ", df.iloc[index3])

#Competidor mas joven que ha recibido oro USA y CAN
NOC3=["USA","CAN"]
df_joven=df[df["NOC"].isin(NOC3)]

df_j_oro=df_joven[df_joven["Medal"]=="Gold"]
index4=df_j_oro["Age"].idxmin()
print("Competidor mas joven que ha recibido oro USA y CAN: ", df.iloc[index4])

#Atleta mas condecorado para los NOC´s USA, CHN y RUS
NOC4=["USA","CHN", "RUS"]
df_condecorado=df_medallas[df_medallas["NOC"].isin(NOC4)]

index5=df_condecorado["Name"].value_counts().idxmax()
print("Atleta mas condecorado para los NOC´s USA, CHN y RUS: ", index5)

condecoradoN=df[df["Name"]==index5]
condecoradoN.to_csv("condecoradoN.csv")

```
