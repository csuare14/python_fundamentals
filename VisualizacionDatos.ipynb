```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("athlete_events-1.csv")

#PARTE 1

#Limpieza datos
df=df[df["Medal"].notnull()]
medallas_pais=df["Team"].value_counts()

#Grafico barras
medallas_pais.head(10).plot(kind="bar", color="darkorange")
plt.suptitle("Pases con ms medallas")
plt.xlabel("Pas")
plt.ylabel("Medallas")
plt.show()


#PARTE 2
ul_ao=df["Year"].max()
df_ul=df[df["Year"]==ul_ao]
medallas_ul_ao=df_ul["Team"].value_counts()

#Grafico torta
medallas_ul_ao.head(10).plot(kind="pie", shadow=True, autopct="%1.1f%%")
plt.suptitle(f"Pases con ms medallas del ao {ul_ao}")
plt.show()
```
